```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder

from sklearn.model_selection import StratifiedKFold, cross_val_predict, cross_val_score
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
import lightgbm as lgb
from sklearn import metrics

import os
import pickle
```

```python
train = pd.read_csv("../data/train.csv")
test = pd.read_csv("../data/test.csv")

#sample_sub = pd.read_csv("../data/sample_submission_fBo3EW5.csv")
print(train.shape, test.shape)
train.head()
```

```python
def eval_top(y, preds):
    y = y.values
    n = int(len(preds) * 0.2)
    indices = np.argsort(preds[:, 1])[::-1][:n]
    #print(len(indices))
    return sum(y[indices])/sum(y)

def cv_oof_predictions(estimator, X, y, cvlist, **est_kwargs):
    preds = np.zeros(len(y)) #Initialize empty array to hold prediction
    scores = []
    ests = []
    for tr_index, val_index in cvlist:
        X_tr , X_val = X.iloc[tr_index], X.iloc[val_index]
        y_tr , y_val = y.iloc[tr_index], y.iloc[val_index]
        est = estimator.set_params(**est_kwargs)
        #print(X_tr.shape, X_val.shape, y_tr.shape, y_val.shape)
        est.fit(X_tr, y_tr, eval_set = [(X_tr, y_tr), (X_val, y_val)], eval_metric='logloss',
               early_stopping_rounds=20, verbose=50) #Might need to change this depending on estimator
        preds[val_index] = est.predict_proba(X_val)[:,1]
        score = eval_top(y_val, est.predict_proba(X_val))
        print(score)
        scores.append(score)
        ests.append(est)
        #break
    print(np.mean(y), np.mean(scores))
    return ests, preds, scores
```

```python
#Premature closing of growth accounts
growth_preclose_feats = ['RD_PREM_CLOSED_PREVQ1' ,'FD_PREM_CLOSED_PREVQ1']

#Premature loan closure
loan_preclose_feats = [f for f in train.columns if ('_PREM_CLOSED' in f) & (f not in growth_preclose_feats)]

#Closure of growth accounts in last quarter
growth_close_feats = ['RD_CLOSED_PREVQ1', 'FD_CLOSED_PREVQ1']

#CLosure of growth accounts in last year
growth_close_feats2 = ['DEMAT_CLOSED_PREV1YR', 'SEC_ACC_CLOSED_PREV1YR']

#Closure of loan accounts in last quarter
loan_close_feats = train.filter(regex='(?<!PREM)(_Closed_PrevQ1|_CLOSED_PREVQ1)').columns.tolist() + ['CC_CLOSED_PREVQ1']
print(len(loan_close_feats))

#Live growth accounts
growth_accounts_live = ['DEMAT_TAG_LIVE', 'SEC_ACC_TAG_LIVE', 'MF_TAG_LIVE', 'RD_TAG_LIVE', 'FD_TAG_LIVE']
print(len(growth_accounts_live))

#Live loan accounts
loan_accounts_live = train.filter(regex='(_LIVE|live)').columns.tolist()
loan_accounts_live = [f for f in loan_accounts_live if f not in growth_accounts_live]
print(len(loan_accounts_live))

#Loan flags
loan_flags = train.filter(regex='_DATE$').columns
print(len(loan_flags))

#CC_PREM_CLOSED_PREVQ1; EDU_PREM_CLOSED_PREVQ1; TL_PREM_CLOSED_PREVQ1; TWL_PREM_CLOSED_PREVQ1
#CE_PREM_CLOSED_PREVQ1; CV_PREM_CLOSED_PREVQ1; AGRI_PREM_CLOSED_PREVQ1
```

```python
#New features - Whether closed any account prematuraly
train['acc_prem_close'] = train[loan_preclose_feats].fillna(0).apply(lambda row: np.any(row), axis=1)
test['acc_prem_close'] = test[loan_preclose_feats].fillna(0).apply(lambda row: np.any(row), axis=1)

print(train.groupby('acc_prem_close')['Responders'].mean())
print(train.groupby('acc_prem_close')['Responders'].size())
```

```python
#How many loans closed prematurely
train['acc_prem_close_sum'] = train[loan_preclose_feats].fillna(0).apply(lambda row: sum(row), axis=1).clip(0,2)
test['acc_prem_close_sum'] = test[loan_preclose_feats].fillna(0).apply(lambda row: sum(row), axis=1).clip(0,2)

print(train.groupby('acc_prem_close_sum')['Responders'].mean())
print(train.groupby('acc_prem_close_sum')['Responders'].size())
```

```python
base_feats = ['EOP_prev1', 'EOP_prev3', 'EOP_prev4', 'EOP_prev6', 'vintage', 'CR_AMB_Prev1']
train[base_feats].describe()
```

```python
loan_flags = [f for f in loan_flags if f not in ['AL_DATE']]
train[loan_flags].describe()
```

```python
train[loan_flags] = train[loan_flags].fillna(0)
train['loan_flag_sum1'] = train[loan_flags].apply(lambda row: sum([r==1 for r in row]), 
                                                  axis=1)
train['loan_flag_sum2'] = train[loan_flags].apply(lambda row: sum([r==2 for r in row]), 
                                                  axis=1)

test[loan_flags] = test[loan_flags].fillna(0)
test['loan_flag_sum1'] = test[loan_flags].apply(lambda row: sum([r==1 for r in row]), 
                                                  axis=1)
test['loan_flag_sum2'] = test[loan_flags].apply(lambda row: sum([r==2 for r in row]), 
                                                  axis=1)
```

```python
new_feats = ['acc_prem_close', 'acc_prem_close_sum', 'loan_close_num', 'growth_close_num',
            'growth2_close_num', 'loan_live', 'growth_live', 'loan_flag_sum1', 'loan_flag_sum2']
train[new_feats].describe()
```

```python
X = train[base_feats]
y = train['Responders']

cvlist = list(StratifiedKFold(n_splits=5, random_state=5).split(X, y))
```

```python
lgb_params = {
    'learning_rate': 0.02,
    'max_depth': -1,
    'num_leaves': 101,
    'n_estimators': 2000,
    'min_child_weight': 11,
    'min_child_samples': 50,
    'subsample':0.8,
    'metric':'logloss',
    'colsample_bytree':0.8,
    'min_sum_hessian_in_leaf':20,
    'reg_lambda': 1,
    'reg_alpha':1,
    #'verbose':1,
}
print("Starting cross -validation")
ests, preds, scores = cv_oof_predictions(lgb.LGBMClassifier(), X, y, cvlist, **lgb_params)
```

```python
lgb_params = {
    'learning_rate': 0.02,
    'max_depth': -1,
    'num_leaves': 151,
    'n_estimators': 2000,
    'min_child_weight': 11,
    'min_child_samples': 10,
    'subsample':0.8,
    'metric':'logloss',
    'colsample_bytree':0.8,
    'min_sum_hessian_in_leaf':20,
    'reg_lambda': 1,
    'reg_alpha':1,
    #'verbose':1,
}
ests, preds, scores = cv_oof_predictions(lgb.LGBMClassifier(), train[base_feats + new_feats], y, cvlist, **lgb_params)
```

```python
cat_cols = train.select_dtypes(['object']).columns
for col in cat_cols:
    lb = LabelEncoder()
    lb.fit(pd.concat([train[col], test[col]]).astype(str).fillna('-1'))
    train[col] = lb.transform(train[col].astype(str).fillna('-1'))
    test[col] = lb.transform(test[col].astype(str).fillna('-1'))
```

```python
train[all_acc_feats] = train[all_acc_feats].fillna(0)
test[all_acc_feats] = test[all_acc_feats].fillna(0)
```
