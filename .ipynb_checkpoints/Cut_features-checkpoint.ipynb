```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder

from sklearn.model_selection import StratifiedKFold, cross_val_predict, cross_val_score
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
import lightgbm as lgb
from sklearn import metrics
```

```python
train = pd.read_csv("../data/train.csv")
test = pd.read_csv("../data/test.csv")

#sample_sub = pd.read_csv("../data/sample_submission_fBo3EW5.csv")
print(train.shape, test.shape)
train.head()
```

```python
cat_cols = train.select_dtypes(['object']).columns
for col in cat_cols:
    lb = LabelEncoder()
    lb.fit(pd.concat([train[col], test[col]]).astype(str).fillna('-1'))
    train[col] = lb.transform(train[col].astype(str).fillna('-1'))
    test[col] = lb.transform(test[col].astype(str).fillna('-1'))
```

```python
#Some feature engineering
#Exponential weighted average of last 6 EOP
for df in train, test:
    df['EOP_exp'] =    df['EOP_prev1'] + 0.5*(
                            df['EOP_prev2'] + 0.5* (
                                df['EOP_prev3']  + 0.5* (
                                    df['EOP_prev4']  +   0.5* (
                                        df['EOP_prev5'] + 0.5 * df['EOP_prev6']))))

    df['CR_AMB_Drop_Build_exp'] =    df['CR_AMB_Drop_Build_1'] + 0.5*(
                                        df['CR_AMB_Drop_Build_2'] + 0.5* (
                                            df['CR_AMB_Drop_Build_3']  + 0.5* (
                                                df['CR_AMB_Drop_Build_4']  +   0.5* (
                                                    df['CR_AMB_Drop_Build_5'] ))))
```

```python
import pickle
with open("../utility/reduced_feats.pkl", "rb") as f:
    red_feats = pickle.load(f)
```

```python
X = train[red_feats]
y = train['Responders']

cvlist = list(StratifiedKFold(n_splits=5, random_state=5).split(X, y))
```

```python
def eval_top(y, preds):
    y = y.values
    n = int(len(preds) * 0.2)
    indices = np.argsort(preds[:, 1])[::-1][:n]
    #print(len(indices))
    return sum(y[indices])/sum(y)

def cv_oof_predictions(estimator, X, y, cvlist, **est_kwargs):
    preds = np.zeros(len(y)) #Initialize empty array to hold prediction
    scores = []
    ests = []
    for tr_index, val_index in cvlist:
        X_tr , X_val = X.iloc[tr_index], X.iloc[val_index]
        y_tr , y_val = y.iloc[tr_index], y.iloc[val_index]
        est = estimator.set_params(**est_kwargs)
        #print(X_tr.shape, X_val.shape, y_tr.shape, y_val.shape)
        est.fit(X_tr, y_tr, eval_set = [(X_tr, y_tr), (X_val, y_val)], eval_metric='logloss',
               early_stopping_rounds=20, verbose=50) #Might need to change this depending on estimator
        preds[val_index] = est.predict_proba(X_val)[:,1]
        score = eval_top(y_val, est.predict_proba(X_val))
        print(score)
        scores.append(score)
        ests.append(est)
        #break
    print(np.mean(y), np.mean(scores))
    return ests, preds, scores
```

```python
lgb_params = {
    'learning_rate': 0.02,
    'max_depth': -1,
    'num_leaves': 251,
    'n_estimators': 2000,
    'min_child_weight': 11,
    'min_child_samples': 50,
    'subsample':0.8,
    'metric':'logloss',
    'colsample_bytree':0.8,
    'min_sum_hessian_in_leaf':20,
    'reg_lambda': 1,
    'reg_alpha':1,
    #'verbose':1,
}
print("Starting cross -validation")
ests, preds, scores = cv_oof_predictions(lgb.LGBMClassifier(), X, y, cvlist, **lgb_params)
```

```python
#Lets club all columns related to customer transaction and try to visualize groups
credit_cols = train.filter(like='_C_').columns.tolist()
debit_cols = train.filter(like='_C_').columns.tolist()
credit_cols2 = train.filter(like='_CR_').columns.tolist()
debit_cols2 = train.filter(like='_DR_').columns.tolist()
cashwit_cols = train.filter(like='_CW_').columns.tolist()
cashdep_cols = train.filter(like='_CASH_Dep').columns.tolist()

transaction_cols = credit_cols + debit_cols + credit_cols2 + debit_cols2 + cashwit_cols + cashdep_cols
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
#from sklearn.manifold import TSNE

transac_data = train[transaction_cols]
transac_data.fillna(np.nanmedian(transac_data), inplace=True)
print(transac_data.shape, transac_data.isnull().sum().sum())
```

```python
pca = PCA(100)
pca.fit(StandardScaler().fit_transform(transac_data))
plt.plot(pca.explained_variance_ratio_.cumsum())
plt.show()
```

```python
X_transformed = pd.concat([pd.DataFrame(PCA(4).fit_transform(StandardScaler().fit_transform(transac_data))), train['Responders']], axis=1)
#fig, ax = plt.subplots(figsize=())
sns.pairplot(X_transformed, hue='Responders', vars=[0,1,2,3], plot_kws={'alpha':0.2})
plt.show()
```

```python
import os
if os.path.exists("../utility/transactions_tsne.pkl"):
    with open("../utility/transactions_tsne.pkl", "rb") as f:
        X_tsne = pickle.load(f)
else:
    from MulticoreTSNE import MulticoreTSNE as TSNE
    tsne = TSNE(n_jobs=8, verbose=1)
    X_tsne = tsne.fit_transform(StandardScaler().fit_transform(transac_data))
    with open("../utility/transactions_tsne.pkl", "wb") as f:
        pickle.dump(X_tsne, f)
```

```python
plt.figure(figsize=(12,10))
palette = sns.color_palette('Set2', 2)
colors = [palette[x] if x >= 0 else (0.0, 0.0, 0.0) for x in train.Responders]
plt.scatter(X_tsne[:,0], X_tsne[:,1], color=colors, alpha=0.2)
plt.show()
```

```python
#Cluster based on transactions
import time
def plot_clusters(data, X_tsne, algorithm, args, kwds):
    start_time = time.time()
    labels = algorithm(*args, **kwds).fit_predict(data)
    end_time = time.time()
    palette = sns.color_palette('deep', np.unique(labels).max() + 1)
    colors = [palette[x] if x >= 0 else (0.0, 0.0, 0.0) for x in labels]
    plt.scatter(X_tsne.T[0], X_tsne.T[1], c=colors, alpha=0.1)
    frame = plt.gca()
    frame.axes.get_xaxis().set_visible(False)
    frame.axes.get_yaxis().set_visible(False)
    plt.title('Clusters found by {}'.format(str(algorithm.__name__)), fontsize=24)
    plt.text(-0.5, 0.7, 'Clustering took {:.2f} s'.format(end_time - start_time), fontsize=14)
    plt.show()
```
