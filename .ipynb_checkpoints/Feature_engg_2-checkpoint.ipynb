```python
from sklearn.metrics import make_scorer
def eval_top(y, preds):
    y = np.array(y)
    n = int(len(preds) * 0.2)
    if np.ndim(preds) == 1:
        indices = np.argsort(preds)[::-1][:n]
    else:
        indices = np.argsort(preds[:, 1])[::-1][:n]
    return sum(y[indices])/sum(y)


eval_sklearn = make_scorer(eval_top, needs_proba=True)

def eval_xgb(preds, dtrain):
    labels = dtrain.get_label()
    eval_score = eval_top(labels, preds)
    return [('eval', eval_score)]

def eval_lgb(actuals, preds):
    return 'eval', eval_top(actuals, preds), True
```

```python
def predict_proba_corr(self, X):
    preds = self.predict_proba(X)[:,1]
    d0 = 0.5
    d1 = 1 - d0
    r0 = np.mean(preds)
    r1 = 1 - r0
    gamma_0 = r0/d0
    gamma_1 = r1/d1
    return gamma_1*preds/(gamma_1*preds + gamma_0*(1 - preds))
```

```python

def cv_oof_predictions(estimator, X, y, cvlist, **est_kwargs):
    preds = np.zeros(len(y)) #Initialize empty array to hold prediction
    scores = []
    ests = []
    for tr_index, val_index in cvlist:
        X_tr , X_val = X.iloc[tr_index], X.iloc[val_index]
        y_tr , y_val = y.iloc[tr_index], y.iloc[val_index]
        est = estimator.set_params(**est_kwargs)
        #print(X_tr.shape, X_val.shape, y_tr.shape, y_val.shape)
        est.fit(X_tr, y_tr, eval_set = [(X_tr, y_tr), (X_val, y_val)], eval_metric=eval_lgb,
               early_stopping_rounds=100, verbose=50,
               #sample_weight = preds_rfc[tr_index]
               ) #Might need to change this depending on estimator
        preds[val_index] = est.predict_proba(X_val)[:,1]
        score = eval_top(y_val, est.predict_proba(X_val))
        print(score)
        scores.append(score)
        ests.append(est)
        #break
    print(np.mean(y), np.mean(scores))
    return ests, preds, scores
```

```python
#Lets encode all categoricals with likelihood
#Writing cv method completely in pandas
def cv_feat_pd1col(data_df, eval_col, target_col, cvlist, func, thresh=3):
    cv_vals = np.ones(len(data_df)) * np.nan #Initialize
    for tr_index, val_index in cvlist:
        tr_func_dict = data_df.loc[tr_index].groupby(eval_col)[target_col].apply(lambda x: 
                                                    func(x) if len(x) >= thresh  else np.nan).to_dict()
        #print(tr_func_dict)
        cv_vals[val_index] = data_df.loc[val_index, eval_col].apply(lambda row: tr_func_dict[row]
                                                                     if row in tr_func_dict
                                                                     else func(data_df.loc[tr_index, target_col]))
    return cv_vals

def cv_feat_pdcols(data_df, eval_cols, target_col, cvlist, func, thresh=3):
    cv_vals = np.ones(len(data_df)) * np.nan #Initialize
    X = data_df.copy()
    X['eval_records'] = data_df[eval_cols].to_records(index=False)
    return cv_feat_pd1col(X, 'eval_records', target_col, cvlist, func, thresh)
```

```python
def cust_obj(y, preds):
    preds = 1. / (1. + np.exp(-preds))
    indices = np.argsort(preds)[::-1][int(0.2*len(preds)):]
    grad = preds - y
    grad[indices] = grad[indices] * 0.001
    hess = preds * (1. - preds)
    hess[indices] = hess[indices] * 0.001
    return grad, hess
```

```python
lgb_params = {
    'learning_rate': 0.018,
    'max_depth': -1,
    'num_leaves': 255,
    'n_estimators': 2000,
    #'min_child_weight': 11,
    #'min_child_samples': 200,
    'subsample':0.95,
    'colsample_bytree':0.5,
    #'min_sum_hessian_in_leaf':20,
    #'reg_lambda': 1,
    #'reg_alpha':1,
    #'is_unbalance':True,
    #'verbose':1,
}
print("Starting cross -validation")
ests, preds, scores = cv_oof_predictions(lgb.LGBMClassifier(), X, y, cvlist, **lgb_params)
```
