```python
import pandas as pd
import pickle
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_names = ['1509937426.604894', '1509942407.685226', '1509943531.061385','1509945689.4197109', 
              '1509947860.395604', '1509949617.7284522', '1509950708.115658','1509956763.777301', 
              '1509958880.3908072', '1509962237.835281', '1509964880.707708','1509967065.004432', 
              '1509969854.118706', '1509971168.6961129', '1509975148.758666','1509975874.420778', 
              '1509976817.403819', '1509978880.332333', '1509980509.988769','1509982172.316803']
```

```python
list_of_dicts = []
for file in file_names:
    path = os.path.join("../utility/", file)
    print(path)
    with open(path, "rb") as f:
        list_of_dicts.append(pickle.load(f))
```

```python
train_oof_preds = pd.DataFrame(columns=range(20))
test_preds = pd.DataFrame(columns=range(20))
for f in range(len(list_of_dicts)):
    train_oof_preds[f] = list_of_dicts[f]["train_preds"]
    test_preds[f] = list_of_dicts[f]["test_preds"]
train_oof_preds.shape, test_preds.shape
```

```python
for f in range(20):
    plt.figure()
    plt.hist(train_oof_preds[f], bins = 100, color = 'orange')
    plt.hist(test_preds[f],bins = 100, alpha = 0.5)
    plt.show()
```

```python
train_oof_preds["avg_oof_preds"] = np.mean(train_oof_preds, axis=1)
test_preds["avg_preds"] = np.mean(test_preds, axis=1)
plt.figure()
plt.hist(train_oof_preds["avg_oof_preds"], bins = 100)
plt.hist(test_preds["avg_preds"], bins = 100, alpha = 0.5)
plt.show()
```

```python
def eval_top(y, preds):
    y = np.array(y)
    n = int(len(preds) * 0.2)
    if np.ndim(preds) == 1:
        indices = np.argsort(preds)[::-1][:n]
    else:
        indices = np.argsort(preds[:, 1])[::-1][:n]
    return sum(y[indices])/sum(y)
```

```python
train = pd.read_csv("../../input/train.csv")
eval_top(train["Responders"], train_oof_preds["avg_oof_preds"])
```

```python
train_oof_preds.to_csv("../utility/train_oof_lgb_opt.csv", compression="gzip")
test_preds.to_csv("../utility/test_pred_lgb_opt.csv", compression="gzip")
```
