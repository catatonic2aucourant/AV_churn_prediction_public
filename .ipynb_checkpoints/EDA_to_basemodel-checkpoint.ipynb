```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn

from sklearn.preprocessing import LabelEncoder

from sklearn.model_selection import StratifiedKFold, cross_val_predict, cross_val_score
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
import lightgbm as lgb
from sklearn import metrics
```

```python
train = pd.read_csv("../data/train.csv")
test = pd.read_csv("../data/test.csv")

#sample_sub = pd.read_csv("../data/sample_submission_fBo3EW5.csv")
print(train.shape, test.shape)
train.head()
```

```python
train.Responders.mean()
```

```python
cat_cols = train.select_dtypes(['object']).columns
for col in cat_cols:
    lb = LabelEncoder()
    lb.fit(pd.concat([train[col], test[col]]).astype(str).fillna('-1'))
    train[col] = lb.transform(train[col].astype(str).fillna('-1'))
    test[col] = lb.transform(test[col].astype(str).fillna('-1'))
```

```python
feats = [f for f in train.columns if f not in ['UCIC_ID', 'Responders']] 
X = train[feats].fillna(-1)
y = train['Responders']

cvlist = list(StratifiedKFold(n_splits=5).split(X, y))
```

```python
def eval_top(y, preds):
    y = y.values
    n = int(len(preds) * 0.2)
    indices = np.argsort(preds[:, 1])[::-1][:n]
    #print(len(indices))
    return sum(y[indices])/sum(y)
```

```python
rfc = RandomForestClassifier(n_estimators=500, verbose=1, n_jobs=-1, max_depth=9, 
                            min_samples_leaf=20)
preds = cross_val_predict(rfc, X, y, cv=cvlist, verbose=1, method='predict_proba')
```

```python
eval_sklearn = metrics.make_scorer(eval_top, needs_proba=True)

#LightGBM
def eval_lgb(actuals, preds):
    return 'gini', eval_top(actuals, preds), True

eval_top(y, preds)
```

```python
def cv_oof_predictions(estimator, X, y, cvlist, **est_kwargs):
    preds = np.zeros(len(y)) #Initialize empty array to hold prediction
    scores = []
    ests = []
    for tr_index, val_index in cvlist:
        X_tr , X_val = X.iloc[tr_index], X.iloc[val_index]
        y_tr , y_val = y.iloc[tr_index], y.iloc[val_index]
        est = estimator.set_params(**est_kwargs)
        #print(X_tr.shape, X_val.shape, y_tr.shape, y_val.shape)
        est.fit(X_tr, y_tr, eval_set = [(X_tr, y_tr), (X_val, y_val)], eval_metric='logloss',
               early_stopping_rounds=20, verbose=50) #Might need to change this depending on estimator
        preds[val_index] = est.predict_proba(X_val)[:,1]
        score = eval_top(y_val, est.predict_proba(X_val))
        print(score)
        scores.append(score)
        ests.append(est)
        #break
    print(np.mean(y), np.mean(scores))
    return ests, preds, scores
```

```python
lgb_params = {
    'learning_rate': 0.02,
    'max_depth': -1,
    'num_leaves': 251,
    'n_estimators': 2000,
    'min_child_weight': 11,
    'min_child_samples': 50,
    'subsample':0.8,
    'metric':'logloss',
    'colsample_bytree':0.8,
    'min_sum_hessian_in_leaf':20,
    'reg_lambda': 1,
    'reg_alpha':1,
    #'verbose':1,
}
print("Starting cross -validation")
ests, preds, scores = cv_oof_predictions(lgb.LGBMClassifier(), X, y, cvlist, **lgb_params)
```

```python
#Some feature engineering
#Exponential weighted average of last 6 EOP
for df in train, test:
    df['EOP_exp'] =    df['EOP_prev1'] + 0.5*(
                            df['EOP_prev2'] + 0.5* (
                                df['EOP_prev3']  + 0.5* (
                                    df['EOP_prev4']  +   0.5* (
                                        df['EOP_prev5'] + 0.5 * df['EOP_prev6']))))

    df['CR_AMB_Drop_Build_exp'] =    df['CR_AMB_Drop_Build_1'] + 0.5*(
                                        df['CR_AMB_Drop_Build_2'] + 0.5* (
                                            df['CR_AMB_Drop_Build_3']  + 0.5* (
                                                df['CR_AMB_Drop_Build_4']  +   0.5* (
                                                    df['CR_AMB_Drop_Build_5'] ))))
```

```python
def plot_importance(forest, feats, n_best=50):
    importances = forest.feature_importances_
    std = np.std([tree.feature_importances_ for tree in forest.estimators_],
                 axis=0)
    indices = np.argsort(importances)[::-1]
    n = len(feats)
    # Print the feature ranking
    print("Feature ranking:")

    for f in range(n):
        print("{} feature {} ({})".format(feats[indices[f]], indices[f], importances[indices[f]]))

    # Plot the feature importances of the forest
    plt.figure(figsize=(14, 12))
    plt.title("Feature importances")
    plt.bar(range(n), importances[indices],
           color="r", yerr=std[indices], align="center")
    plt.xticks(range(n), np.array(feats)[indices], rotation=90)
    plt.xlim([-1, n_best])
    plt.show()
```
